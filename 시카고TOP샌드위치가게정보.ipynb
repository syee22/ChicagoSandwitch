```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.request import Request
import pandas as pd
import numpy as np
```

```python
url='https://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago'
```

```python
# html = urlopen(url) #403번 에러 발생(부정접근으로 차단-알려진 agent를 헤더에 추가해서 전달하면 됨)
# 헤더를 추가하려면 일반적으로 Request(get_url,headers={key:value})
import ssl
context = ssl._create_unverified_context()
req = Request(url,headers={'User-Agent':'Mozilla/5.0'})
html=urlopen(req,context=context)

h = BeautifulSoup(html,'html.parser')
```

```python
sandwitch=h.findAll('div',{'class':'sammyListing'})
len(sandwitch)
```

```python
url_base='https://www.chicagomag.com/'
subs=[]
urls=[]
for i in range(len(sandwitch)):
    if url_base in sandwitch[i].find('a')['href']:
        subs.append(sandwitch[i].find('a')['href'])
        urls.append(subs[i])
    else:
        subs.append(sandwitch[i].find('a')['href'])
        urls.append(url_base+subs[i])
    
```

```python
for i in range(50):
    print(urls[i])
```

```python
com_df=pd.DataFrame({'BEST menu':[],'price':[],'tel':[],'Address':[],'comp_url':[]})
infos=[]
company_url=[]
best_m=[]
tel=[]
price=[]
addr=[]
for i in range(0,len(urls)):
    url=urls[i]
    context = ssl._create_unverified_context()
    req = Request(url,headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'})
    html=urlopen(req,context=context)
    h = BeautifulSoup(html,'html.parser')
    
    best_m.append((h.find('div',{'class':'art-head h60'}).text)[3:])
    infos.append(h.find('p','addy').find('em').text)
    price.append(infos[i].split(',')[0].split(maxsplit=1)[0])
    addr.append(infos[i].split(',')[0].split(maxsplit=1)[1])
    
    try:
        if infos[i].split(',')[-1][4]=='-' :
            tel.append(infos[i].split(',')[-1])
        elif infos[i].split(',')[1][4]=='-' :
            tel.append(infos[i].split(',')[1])
        elif infos[i].split(',')[2][4]=='-' :
            tel.append(infos[i].split(',')[2])

    except:
        tel.append('no data')
    try:
        company_url.append(urls[i])
    except:
        company_url.append('no data')
    
        
    df=pd.DataFrame({'BEST menu':best_m,'price':price,
                     'tel':tel,'Address':addr,
                     'comp_url':company_url})
    fin=pd.concat([df,com_df],ignore_index=False)
```
